```python
ss = []
for t in [(1984, 1986.5),
          (1990.3, 1992.1),
          (1994.6, 1995.6),
          (1996.1,1996.5),
          (1997,1997.6),
          (1997.75,1998.15),
          (1998.25,1999.35),
          (2000.55,2000.8),
          (2000.9 ,2001.25),
          (2001.6 ,2002),
          (2002.5 ,2003)]:

    # for each pair I convert the year to datetime (easy)
    # then I convert the fraction of year and convert it to days and convert that to a delta-time
    # finally I add that deltatime to the year datetime.
    print(t)
    a = pd.to_datetime(int(t[0]), format="%Y") + pd.Timedelta(days=((t[0] - int(t[0])) * 365.25))
    b = pd.to_datetime(int(t[1]), format="%Y") + pd.Timedelta(days=((t[1] - int(t[1])) * 365.25))

    ss.append((a, b))
```

## removing the first 10 smallest samples after dividing by mean  as suggested in Section3 para 3


```python



mag_thresholds = [2.0, 2.5, 3.0, 3.5, 4.0]
markers = ['o', 'x', '*', 's', 'D']

cb_list = []
cs_list = []

# Loop through the MAG thresholds and process each dataset
for mag_threshold in mag_thresholds:

    # x = np.log10((eq_final[eq_final['MAG'] > mag_threshold]['gaps']))
    x = (eq_final[eq_final['MAG'] > mag_threshold]['gaps'])

    # Calculate cumulative distribution and center bin values
    def hist_bin(x, bin):
        hist, bins = np.histogram(x, bins= bin)
        cb = (bins[:-1] + bins[1:]) / 2
        cs = np.cumsum(hist) / np.sum(hist)
        return cb, cs

    def c_mean(x):
        fmx = np.mean(x)
        x = x / fmx
        x1 = x[x > 0.01]
        smx = np.mean(x1)
        x1 = x1 / smx
        cb, cs = hist_bin(x1, bin =1900)
        return cb, cs

    cb, cs = c_mean(x)


    cb_list.append(cb)
    cs_list.append(cs)


plt.figure(figsize=(10, 6))
for cb, cs, marker in zip(cb_list, cs_list, markers):
    plt.plot(cb, cs, marker=marker)

plt.xlabel('time_gaps (dimensionless)')
plt.ylabel('probability (dimensionless)')
plt.title('Rescaled inter-event-time cumulative distributions for Southern California m=0.01. For Different MAG Thresholds')
plt.legend([f'MAG > {mag}' for mag in mag_thresholds], loc='lower right')
plt.grid(True)
plt.xscale('log')
plt.text(0.5, -0.3, "cumulative plot of the counts for various threshold: to see \n if they follow the same trend or the law of scaling as \n as x axis is normalized", ha='center', fontsize=12, transform=plt.gca().transAxes)

plt.show()
```

```python


mag_thresholds = [2.0, 2.5, 3.0, 3.5, 4.0]
p_mat = np.zeros((len(mag_thresholds), len(mag_thresholds)))
for i in range(len(mag_thresholds)):
    xi = eq_final[eq_final['MAG'] > mag_thresholds[i]]['gaps']
    print(len(xi))
    bin_size = int( 0.01* len(xi))
    cbi, chisti = c_mean(xi, bin = bin_size, m = 0.01)


    for j in range(len(mag_thresholds)):
        if i == j:
            p_mat[i, j] = 1
        elif i < j:

            xj = eq_final[eq_final['MAG'] > mag_thresholds[j]]['gaps']
            bin_size = int( 0.01* len(xj))
            cbj, chistj = c_mean(xj,bin = bin_size, m= 0.01)




            ks_stat, p_value = scipy.stats.kstest(chisti, chistj)


            p_mat[i, j] = p_value
        else:
            p_mat[i, j] = p_mat[j, i]

plt.imshow(p_mat, cmap='viridis', vmin=0, vmax=1)
plt.colorbar(label=r'$p$-value')
plt.title("KS Test Results for Different MAG Thresholds")
plt.axis('off')


plt.text(0.5, -0.1, "P-value Heatmap for diffrent configration, : helps to understant the kstest of all diffrent conbination in a single picture", ha='center', fontsize=12, transform=plt.gca().transAxes)

plt.show()
print(p_mat)
```

```python
alpha = 0.05
significant_mask = p_mat < alpha

colors = [(1, 1, 1), (1, 0, 0)]  # White to Red
cmap = LinearSegmentedColormap.from_list("Custom_Red_White_Colormap", colors, N=256)
plt.imshow(significant_mask, cmap=cmap, vmin=0, vmax=1)
cbar = plt.colorbar()
cbar.set_ticks([0, 1])
cbar.set_ticklabels(['Non-Significant', 'Significant'])
plt.text(0.5, -0.1, "Statistical Significance Heatmap for threshold of 0.05, any pvalue less than equal to 0.05 (red) is  statistically significant and not by chance", ha='center', fontsize=12, transform=plt.gca().transAxes)
plt.show()
```

```python
alpha = 0.001
significant_mask = p_mat < alpha

colors = [(1, 1, 1), (1, 0, 0)]  # White to Red
cmap = LinearSegmentedColormap.from_list("Custom_Red_White_Colormap", colors, N=256)
plt.imshow(significant_mask, cmap=cmap, vmin=0, vmax=1)
cbar = plt.colorbar()
cbar.set_ticks([0, 1])
cbar.set_ticklabels(['Non-Significant', 'Significant'])
plt.text(0.5, -0.1, "Statistical Significance Heatmap for threshold of 0.05, any pvalue less than equal to 0.05 (red) is  statistically significant and not by chance", ha='center', fontsize=12, transform=plt.gca().transAxes)
plt.show()
```

## conlusion
It suggest that at alpha  = 0.05 confidence level only few red box appear(meaning 4 out of 20 chance to reject the null hypothisis) greater chance of statistical significance  to reject null hypothsis holds but at alpha = 0.001 confidence level there are all white boxes so Null cannot be rejected at all.
