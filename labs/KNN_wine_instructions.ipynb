```python
import scipy as sp
sp.spatial.distance?
#distance.#pdist?
```

```python
import numpy as np
from scipy.spatial.distance import pdist, squareform

# Create a set of sample points
points = np.array([[1, 2], [3, 4], [5, 6], [7, 8]])

# Calculate pairwise Euclidean distances using pdist
distance_vector = pdist(points, metric='euclidean')

# Convert the distance vector into a square-form distance matrix
distance_matrix = squareform(distance_vector)

# Print the distance vector and the resulting square-form distance matrix
print("Pairwise Euclidean Distances (Vector):")
print(distance_vector)
print("\nPairwise Euclidean Distance Matrix:")
print(distance_matrix)
```

# data exploration

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,wines["quality"], random_state =42)


```

# THIS WILL AID VISUALIZATION

```python
import sklearn.manifold as man #t distributed stochastic neighbourhood embedding
tsne_wines = man.TSNE(n_components=2, perplexity=30.0, early_exaggeration=90.0,
                      random_state=302).fit(X)
wine_proj = tsne_wines.fit_transform(X)
wine_proj_test = tsne_wines.fit_transform(X_test)
```

```python
import seaborn as sns
_, ax = plt.subplots()


ax.scatter(wine_proj.T[0], wine_proj.T[1], s=10, c = wines["quality"]/3);
ax.set_title("true labels")
plt.xlabel("tsny meaningless axis 1")
plt.ylabel("tsny meaningless axis 2")
```
